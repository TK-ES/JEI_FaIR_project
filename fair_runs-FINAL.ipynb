```python
def sector_calc(specie2):
    #sectors
    pickles_to_add = []
    geos = ['MAGICC AFOLU','MAGICC AFOLU|Agricultural Waste Burning', 'MAGICC AFOLU|Agriculture', 'MAGICC AFOLU|Forest Burning','MAGICC AFOLU|Grassland Burning', 
             'MAGICC AFOLU|Peat Burning', 'MAGICC Fossil and Industrial', 'MAGICC Fossil and Industrial|Aircraft',
            'MAGICC Fossil and Industrial|Energy Sector', 'MAGICC Fossil and Industrial|Industrial Sector',
            'MAGICC Fossil and Industrial|International Shipping','MAGICC Fossil and Industrial|Residential Commercial Other', 
           'MAGICC Fossil and Industrial|Solvents Production and Application','MAGICC Fossil and Industrial|Transportation Sector', 'MAGICC Fossil and Industrial|Waste',]
    geos2 = [
        '_'.join(word.lower().replace(' ', '_') for word in geo.split('|')) 
        if index in (0, 6) else
        '_'.join(word.lower().replace(' ', '_') for word in geo.split('|') 
                 if 'magicc afolu' not in word.lower() and 'magicc fossil and industrial' not in word.lower())
        for index, geo in enumerate(geos)
    ]
    specie = specie2.lower().replace(" ", "_").replace("-", "_")
    ssps = ['ssp126', 'ssp245', 'ssp370', 'ssp585']
    rcmip_em_data_og = pd.read_csv("../rcmip_data/rcmip-emissions-annual-means-v5-1-0--OG.csv")#pressed tab
    rcmip_em_data = rcmip_em_data_og.copy(deep=True)
    rcmip_em_data.to_csv("../rcmip_data/rcmip-emissions-annual-means-v5-1-0.csv", index=False)
    locals()[f'{specie}'] = rcmip_em_data[
        (rcmip_em_data["Variable"] == f'Emissions|{specie2}')
    ].copy()
    locals()[f'{specie}_og'] = locals()[f'{specie}'].copy()

    for geo in range(len(geos)):
        var_name = geos2[geo]
        locals()[f'{specie}_minus_{var_name}'] = locals()[f'{specie}_og'].copy()
        rcmip_em_data_og = pd.read_csv("../rcmip_data/rcmip-emissions-annual-means-v5-1-0--OG.csv")#pressed tab
        rcmip_em_data = rcmip_em_data_og.copy(deep=True)
        rcmip_em_data.to_csv("../rcmip_data/rcmip-emissions-annual-means-v5-1-0.csv", index=False)
        locals()[f'{specie}_{var_name}'] = rcmip_em_data[(rcmip_em_data["Variable"] == f'Emissions|{specie2}|{geos[geo]}')].copy()
        print(var_name)

        for geo_row in (locals()[f'{specie}_{var_name}']).index:
            locals()[f'temp_{specie}'] = close_value(locals()[f'{specie}'].index, geo_row)
            locals()[f'{specie}_minus_{var_name}'].loc[locals()[f'temp_{specie}'],locals()[f'{specie}_minus_{var_name}'].columns[2015-1750+7:]] = locals()[f'{specie}_og'].loc[locals()[f'temp_{specie}'],locals()[f'{specie}_og'].columns[2015-1750+7:]] - (((locals()[f'{specie}_{var_name}']).loc[geo_row,(locals()[f'{specie}_{var_name}']).columns[2015-1750+7:]]) - (((locals()[f'{specie}_{var_name}']).loc[geo_row,(locals()[f'{specie}_{var_name}']).columns[1850+7-1750:1900+7-1750]]).mean(axis=0)))
    # for i in (locals()[f'{specie}']).index:
    #     locals()[f'{specie}'].loc[i,locals()[f'{specie}'].columns[2015-1750+7:]] = 0

        rcmip_em_data.update(locals()[f'{specie}_minus_{var_name}'])
        rcmip_em_data.to_csv("../rcmip_data/rcmip-emissions-annual-means-v5-1-0.csv", index=False)
    #    FAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIR
        f1 = FAIR(ch4_method='thornhill2021', ghg_method='meinshausen2020')
        f1.define_time(1750, 2100, 1/12)
        scenarios = ['ssp119', 'ssp126', 'ssp245', 'ssp370', 'ssp434', 'ssp460', 'ssp534-over', 'ssp585']
        f1.define_scenarios(scenarios)
        df = pd.read_csv("../tests/test_data/4xCO2_cummins_ebm3.csv")
        models = df['model'].unique()
        configs = []

        for imodel, model in enumerate(models):
            for run in df.loc[df['model']==model, 'run']:
                configs.append(f"{model}_{run}")
        f1.define_configs(configs)
        species, properties = read_properties()
        f1.define_species(species, properties)

        f1.allocate()
        f1.fill_species_configs()
        fill(f1.species_configs['unperturbed_lifetime'], 10.8537568, specie='CH4')
        fill(f1.species_configs['baseline_emissions'], 19.01978312, specie='CH4')
        fill(f1.species_configs['baseline_emissions'], 0.08602230754, specie='N2O')
        df_volcanic = pd.read_csv('../tests/test_data/volcanic_ERF_monthly_175001-201912.csv', index_col='year')
        f1.fill_from_rcmip()

        # overwrite volcanic
        volcanic_forcing = np.zeros(4201)
        volcanic_forcing[:270*12] = df_volcanic[1750:].squeeze().values
        fill(f1.forcing, volcanic_forcing[:, None, None], specie="Volcanic")  # sometimes need to expand the array

        initialise(f1.concentration, f1.species_configs['baseline_concentration'])
        initialise(f1.forcing, 0)
        initialise(f1.temperature, 0)
        initialise(f1.cumulative_emissions, 0)
        initialise(f1.airborne_emissions, 0)
        df = pd.read_csv("../tests/test_data/4xCO2_cummins_ebm3.csv")
        models = df['model'].unique()

        seed = 1355763

        # remember we rescale sigma
        for config in configs:
            model, run = config.split('_')
            condition = (df['model']==model) & (df['run']==run)
            fill(f1.climate_configs['ocean_heat_capacity'], df.loc[condition, 'C1':'C3'].values.squeeze(), config=config)
            fill(f1.climate_configs['ocean_heat_transfer'], df.loc[condition, 'kappa1':'kappa3'].values.squeeze(), config=config)
            fill(f1.climate_configs['deep_ocean_efficacy'], df.loc[condition, 'epsilon'].values[0], config=config)
            fill(f1.climate_configs['gamma_autocorrelation'], df.loc[condition, 'gamma'].values[0], config=config)
            fill(f1.climate_configs['sigma_eta'], df.loc[condition, 'sigma_eta'].values[0]/np.sqrt(f1.timestep), config=config)
            fill(f1.climate_configs['sigma_xi'], df.loc[condition, 'sigma_xi'].values[0]/np.sqrt(f1.timestep), config=config)
            fill(f1.climate_configs['stochastic_run'], True, config=config)
            fill(f1.climate_configs['use_seed'], True, config=config)
            fill(f1.climate_configs['seed'], seed, config=config)

            seed = seed + 399
        #FAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIRFAIR
        f1.run()
        temperature_data_list = []

        for i, ssp in enumerate(ssps):
            temperature_data = f1.temperature.loc[dict(scenario=ssp, layer=0)]
            temperature_data_list.append(temperature_data)
        to_open = f'{specie}_no_{var_name}_ref1850'
        pickles_to_add.append(to_open)
        with open(to_open, 'wb') as file:
            pickle.dump(temperature_data_list, file)
    return pickles_to_add
```
